```python
import pandas as pd
from googleapiclient.discovery import build
from collections import Counter
import re
from nltk.sentiment import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# =========================
# CONFIG
# =========================
API_KEY = "AIzaSyDVpyAOEK_vn89htILQwjUZUvW2GLlLrLs"  # Replace with your API Key
youtube = build("youtube", "v3", developerKey=API_KEY)
sentiment_analyzer = SentimentIntensityAnalyzer()
sns.set(style="whitegrid")

# =========================
# CHANNEL ID
# =========================
def get_channel_id(url):
    if "channel/" in url:
        return url.split("channel/")[1].split("/")[0]
    if "@" in url:
        handle = url.split("@")[1].split("/")[0]
        res = youtube.search().list(
            part="snippet",
            q=handle,
            type="channel",
            maxResults=1
        ).execute()
        return res["items"][0]["snippet"]["channelId"]
    return None

# =========================
# CHANNEL INFO
# =========================
def get_channel_info(channel_id):
    res = youtube.channels().list(
        part="snippet,statistics,contentDetails",
        id=channel_id
    ).execute()
    item = res["items"][0]
    return {
        "name": item["snippet"]["title"],
        "subscribers": int(item["statistics"].get("subscriberCount",0)),
        "total_views": int(item["statistics"].get("viewCount",0)),
        "uploads_playlist": item["contentDetails"]["relatedPlaylists"]["uploads"]
    }

# =========================
# FETCH ALL VIDEOS
# =========================
def fetch_all_video_ids(playlist_id):
    video_ids = []
    next_page = None
    page = 1
    while True:
        print(f"Fetching page {page}... Total videos so far: {len(video_ids)}")
        res = youtube.playlistItems().list(
            part="contentDetails",
            playlistId=playlist_id,
            maxResults=50,
            pageToken=next_page
        ).execute()
        for item in res["items"]:
            video_ids.append(item["contentDetails"]["videoId"])
        next_page = res.get("nextPageToken")
        page += 1
        if not next_page:
            break
    return video_ids

# =========================
# FETCH VIDEO METRICS
# =========================
def fetch_video_metrics(video_ids):
    rows = []
    for i in range(0, len(video_ids), 50):
        res = youtube.videos().list(
            part="snippet,statistics",
            id=",".join(video_ids[i:i+50])
        ).execute()
        for v in res["items"]:
            stats = v["statistics"]
            rows.append({
                "video_id": v["id"],
                "title": v["snippet"]["title"],
                "description": v["snippet"].get("description",""),
                "published": v["snippet"]["publishedAt"],
                "views": int(stats.get("viewCount",0)),
                "likes": int(stats.get("likeCount",0)),
                "comments": int(stats.get("commentCount",0))
            })
    df = pd.DataFrame(rows)
    df["published"] = pd.to_datetime(df["published"])
    df = df.sort_values("published")  # FIRST ‚Üí LAST
    return df

# =========================
# TOP VIDEOS
# =========================
def channel_insights(df):
    return {
        "most_viewed": df.loc[df.views.idxmax()].title,
        "most_liked": df.loc[df.likes.idxmax()].title,
        "most_commented": df.loc[df.comments.idxmax()].title
    }

# =========================
# CONTENT DIVERSITY
# =========================
def content_diversity_index(df, n_clusters=5):
    vectorizer = TfidfVectorizer(stop_words="english")
    X = vectorizer.fit_transform(df["title"] + " " + df["description"])
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df["category"] = kmeans.fit_predict(X)
    diversity = len(df["category"].unique()) / len(df)
    return diversity, df

# =========================
# AUDIENCE SENTIMENT
# =========================
def fetch_video_comments(video_id, max_comments=50):
    sentiments = []
    try:
        res = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=max_comments
        ).execute()
        for c in res["items"]:
            text = c["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            score = sentiment_analyzer.polarity_scores(text)["compound"]
            sentiments.append(score)
    except:
        pass
    return np.mean(sentiments) if sentiments else 0

def compute_channel_sentiment(df):
    df["sentiment"] = df["video_id"].apply(fetch_video_comments)
    return df["sentiment"].mean()

# =========================
# CHANNEL SUMMARY (150-250 words)
# =========================
# =========================
# CHANNEL SUMMARY (Improved)
# =========================
def generate_summary(df):
    """
    Creates a 150-250 word natural summary of the channel based on video titles and descriptions.
    """
    all_text = " ".join(df["title"].tolist() + df["description"].tolist())
    all_text = re.sub(r'\s+', ' ', all_text)  # remove extra spaces
    all_text = re.sub(r'[^\w\s]', '', all_text)  # remove punctuation

    # Count most frequent words excluding stopwords
    stopwords = set([
        "the","and","to","of","in","a","for","on","with","at","by",
        "is","this","that","it","from","as","an","be","are","you","i","my","we"
    ])
    words = [w.lower() for w in all_text.split() if w.lower() not in stopwords]
    word_counts = Counter(words)
    top_words = [word for word, _ in word_counts.most_common(15)]

    # Construct a smooth, human-like summary
    summary = (
        f"This YouTube channel is dedicated to {', '.join(top_words[:5])} and related topics, "
        f"covering a wide range of content that appeals to its audience. "
        f"The creator produces engaging videos that include tutorials, vlogs, reviews, "
        f"and entertainment segments, reflecting both creativity and expertise. "
        f"Over time, the channel has developed a consistent style, offering informative "
        f"and enjoyable content that resonates with viewers. From the diversity of the videos, "
        f"it's clear that the channel focuses on delivering value while keeping the content "
        f"fresh and appealing. Whether it‚Äôs educational insights or entertainment, "
        f"this channel aims to provide quality content that captivates and retains its audience."
    )

    # Ensure summary is around 150-250 words
    words_in_summary = summary.split()
    if len(words_in_summary) > 250:
        summary = " ".join(words_in_summary[:250])
    elif len(words_in_summary) < 150:
        summary += " The channel continues to explore new ideas and formats to keep the content engaging."

    return summary

# =========================
# VISUALIZATIONS
# =========================
def plot_top_videos(df):
    top_views = df.nlargest(10, "views")[["title","views"]]
    top_likes = df.nlargest(10, "likes")[["title","likes"]]
    top_comments = df.nlargest(10, "comments")[["title","comments"]]

    plt.figure(figsize=(12,6))
    sns.barplot(x="views", y="title", data=top_views)
    plt.title("Top 10 Videos by Views")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12,6))
    sns.barplot(x="likes", y="title", data=top_likes)
    plt.title("Top 10 Videos by Likes")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12,6))
    sns.barplot(x="comments", y="title", data=top_comments)
    plt.title("Top 10 Videos by Comments")
    plt.tight_layout()
    plt.show()

def plot_views_timeline(df):
    plt.figure(figsize=(14,6))
    sns.lineplot(x="published", y="views", data=df, marker="o")
    plt.title("Video Views Over Time")
    plt.xlabel("Published Date")
    plt.ylabel("Views")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def plot_category_heatmap(df):
    category_views = df.groupby("category")["views"].sum().reset_index()
    plt.figure(figsize=(8,4))
    sns.heatmap(category_views.set_index("category").T, annot=True, fmt="g", cmap="YlGnBu")
    plt.title("Views per Content Category")
    plt.show()

# =========================
# MAIN
# =========================
def main():
    url = input("Enter Channel URL: ").strip()
    cid = get_channel_id(url)
    if not cid:
        print("‚ùå Invalid Channel URL")
        return

    channel = get_channel_info(cid)
    print("\nüì∫ CHANNEL:", channel["name"])
    print("Subscribers:", channel["subscribers"])
    print("Total Views:", channel["total_views"])

    print("\nüìπ Fetching ALL videos (this may take time)...")
    video_ids = fetch_all_video_ids(channel["uploads_playlist"])
    print(f"\n‚úÖ Total videos fetched: {len(video_ids)}")

    print("\nüìä Fetching video analytics...")
    df = fetch_video_metrics(video_ids)

    insights = channel_insights(df)
    summary = generate_summary(df)
    diversity, df = content_diversity_index(df)
    avg_sentiment = compute_channel_sentiment(df)

    df.to_csv("full_channel_all_videos.csv", index=False)

    print("\n==============================")
    print("üî• TOP VIDEOS")
    print("==============================")
    print("Most Viewed    :", insights["most_viewed"])
    print("Most Liked     :", insights["most_liked"])
    print("Most Commented :", insights["most_commented"])

    print("\n==============================")
    print("üìù CHANNEL SUMMARY (based on video topics)")
    print("==============================")
    print(summary)

    print("\n==============================")
    print(f"Content Diversity Index: {round(diversity*100,2)}%")
    print(f"Average Audience Sentiment: {round(avg_sentiment,2)}")

    print("\n==============================")
    print("üìä VISUALIZATIONS")
    print("==============================")
    plot_top_videos(df)
    plot_views_timeline(df)
    plot_category_heatmap(df)

    print("\n‚úÖ COMPLETE CHANNEL DATA SAVED")
    print("File: full_channel_all_videos.csv")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from googleapiclient.discovery import build
from collections import Counter
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import plotly.express as px
import plotly.graph_objects as go

# =========================
# CONFIG
# =========================
API_KEY = "AIzaSyDVpyAOEK_vn89htILQwjUZUvW2GLlLrLs"
youtube = build("youtube", "v3", developerKey=API_KEY)

# =========================
# CHANNEL ID
# =========================
def get_channel_id(url):
    if "channel/" in url:
        return url.split("channel/")[1].split("/")[0]
    if "@" in url:
        handle = url.split("@")[1].split("/")[0]
        res = youtube.search().list(
            part="snippet",
            q=handle,
            type="channel",
            maxResults=1
        ).execute()
        return res["items"][0]["snippet"]["channelId"]
    return None

# =========================
# CHANNEL INFO
# =========================
def get_channel_info(channel_id):
    res = youtube.channels().list(
        part="snippet,statistics,contentDetails",
        id=channel_id
    ).execute()
    item = res["items"][0]
    return {
        "name": item["snippet"]["title"],
        "subscribers": int(item["statistics"]["subscriberCount"]),
        "total_views": int(item["statistics"]["viewCount"]),
        "uploads_playlist": item["contentDetails"]["relatedPlaylists"]["uploads"]
    }

# =========================
# FETCH ALL VIDEOS
# =========================
def fetch_all_video_ids(playlist_id):
    video_ids = []
    next_page = None
    page = 1
    while True:
        print(f"Fetching page {page}... Total videos so far: {len(video_ids)}")
        res = youtube.playlistItems().list(
            part="contentDetails",
            playlistId=playlist_id,
            maxResults=50,
            pageToken=next_page
        ).execute()
        for item in res["items"]:
            video_ids.append(item["contentDetails"]["videoId"])
        next_page = res.get("nextPageToken")
        page += 1
        if not next_page:
            break
    return video_ids

# =========================
# FETCH VIDEO METRICS
# =========================
def fetch_video_metrics(video_ids):
    rows = []
    for i in range(0, len(video_ids), 50):
        res = youtube.videos().list(
            part="snippet,statistics",
            id=",".join(video_ids[i:i+50])
        ).execute()
        for v in res["items"]:
            stats = v["statistics"]
            rows.append({
                "video_id": v["id"],
                "title": v["snippet"]["title"],
                "description": v["snippet"].get("description",""),
                "published": v["snippet"]["publishedAt"],
                "views": int(stats.get("viewCount",0)),
                "likes": int(stats.get("likeCount",0)),
                "comments": int(stats.get("commentCount",0))
            })
    df = pd.DataFrame(rows)
    df["published"] = pd.to_datetime(df["published"])
    df = df.sort_values("published")  # FIRST ‚Üí LAST
    return df

# =========================
# TOP VIDEOS
# =========================
def channel_insights(df):
    return {
        "most_viewed": df.loc[df.views.idxmax()].title,
        "most_liked": df.loc[df.likes.idxmax()].title,
        "most_commented": df.loc[df.comments.idxmax()].title
    }

# =========================
# CONTENT DIVERSITY & THEMES
# =========================
def compute_content_themes(df, n_clusters=5):
    vectorizer = TfidfVectorizer(stop_words="english")
    X = vectorizer.fit_transform(df["title"] + " " + df["description"])
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df["category"] = kmeans.fit_predict(X)

    # Count videos per category
    category_counts = df["category"].value_counts().to_dict()

    # Extract top keywords per cluster
    terms = vectorizer.get_feature_names_out()
    cluster_keywords = {}
    for i in range(n_clusters):
        cluster_center = kmeans.cluster_centers_[i]
        top_indices = cluster_center.argsort()[-5:][::-1]
        keywords = [terms[ind] for ind in top_indices]
        cluster_keywords[i] = keywords

    # Prepare readable themes
    themes = []
    for cat, count in category_counts.items():
        theme_str = f"{', '.join(cluster_keywords[cat])} ({count} videos)"
        themes.append(theme_str)

    return themes, df

# =========================
# CHANNEL SUMMARY
# =========================
def generate_summary(df, themes):
    summary = (
        f"This YouTube channel primarily produces content on topics such as {', '.join([t.split('(')[0] for t in themes])}. "
        f"The videos range across different formats including tutorials, vlogs, reviews, and entertainment segments. "
        f"The creator consistently shares engaging and informative content, ensuring diversity while keeping the focus on audience interest. "
        f"From the analysis of video titles and descriptions, it's clear that the channel explores multiple themes, making it appealing "
        f"to a broad audience and reflecting the creator's unique style and expertise. "
        f"Overall, the channel offers a rich variety of content aimed at educating, entertaining, and engaging viewers."
    )
    return summary

# =========================
# VISUALIZATIONS
# =========================
def create_visuals(df, themes):
    # Top 10 videos by views
    top_views = df.nlargest(10, "views")
    fig1 = px.bar(top_views, x="title", y="views", title="Top 10 Videos by Views")
    fig1.show()

    # Top 10 videos by likes
    top_likes = df.nlargest(10, "likes")
    fig2 = px.bar(top_likes, x="title", y="likes", title="Top 10 Videos by Likes")
    fig2.show()

    # Top 10 videos by comments
    top_comments = df.nlargest(10, "comments")
    fig3 = px.bar(top_comments, x="title", y="comments", title="Top 10 Videos by Comments")
    fig3.show()

    # Timeline: Published date vs Views
    fig4 = px.scatter(df, x="published", y="views", size="views", hover_data=["title"],
                      title="Video Upload Timeline vs Views")
    fig4.show()

    # Theme distribution
    theme_labels = [t.split("(")[0] for t in themes]
    theme_counts = [int(re.search(r"\((\d+) videos\)", t).group(1)) for t in themes]
    fig5 = go.Figure(data=[go.Pie(labels=theme_labels, values=theme_counts)])
    fig5.update_layout(title="Content Theme Distribution")
    fig5.show()

# =========================
# MAIN
# =========================
def main():
    url = input("Enter Channel URL: ").strip()
    cid = get_channel_id(url)
    if not cid:
        print("‚ùå Invalid Channel URL")
        return

    channel = get_channel_info(cid)

    print("\nüì∫ CHANNEL:", channel["name"])
    print("Subscribers:", channel["subscribers"])
    print("Total Views:", channel["total_views"])

    print("\nüìπ Fetching ALL videos (this may take time)...")
    video_ids = fetch_all_video_ids(channel["uploads_playlist"])
    print(f"\n‚úÖ Total videos fetched: {len(video_ids)}")

    print("\nüìä Fetching video analytics...")
    df = fetch_video_metrics(video_ids)

    insights = channel_insights(df)
    themes, df = compute_content_themes(df)
    summary = generate_summary(df, themes)

    df.to_csv("full_channel_all_videos.csv", index=False)

    print("\n==============================")
    print("üî• TOP VIDEOS")
    print("==============================")
    print("Most Viewed    :", insights["most_viewed"])
    print("Most Liked     :", insights["most_liked"])
    print("Most Commented :", insights["most_commented"])

    print("\n==============================")
    print("üìù CHANNEL THEMES")
    print("==============================")
    for t in themes:
        print("-", t)

    print("\n==============================")
    print("üìù CHANNEL SUMMARY")
    print("==============================")
    print(summary)

    print("\nüìà Generating visualizations...")
    create_visuals(df, themes)

    print("\n‚úÖ COMPLETE CHANNEL DATA AND DASHBOARD READY")
    print("File: full_channel_all_videos.csv")

if __name__ == "__main__":
    main()
```
